```python
def caluculate_percentage_bounds(data):
    mean = np.mean(data)
    std = np.std(data)
    lower_bound = mean - (2*std)
    upper_bound = mean + (2*std)
    return lower_bound, upper_bound
```

```python
def returns_based_on_days(final_data, num_days):
    returns_d = []
    for i in range(num_days):
        returns_d.append((np.NaN,np.NaN))
    high_low_data = final_data[["high","low"]].values
    for num in range(final_data.shape[0]-num_days):
        high = high_low_data[num][0]
        low = high_low_data[num+num_days][1]
        returns_d.append((high,low))
    returns = [j-i for i,j in returns_d]   
    returns_percent = [((j-i)/i)*num_days for i,j in returns_d]
    return returns, returns_percent
```

```python
final_data["30D - Returns"], final_data["% 30D - Returns"] = returns_based_on_days(final_data, num_days=30)
```

```python
final_data["50D - Returns"], final_data["% 50D - Returns"] = returns_based_on_days(final_data, num_days=50)
```
